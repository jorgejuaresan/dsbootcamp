```python
#Homework No5

import pandas as pd
```

```python
#Import file:
data_ride = pd.read_csv('Resources/hw_5_ride_data.csv')

#data visualitation:
```

```python
#Import file:
data_city = pd.read_csv('Resources/hw_5_city_data.csv')


#data visualitation:
data_city.head()
```

```python
#CITY as a pivot

new_base = pd.merge(data_ride, data_city, on = 'city')

new_base.head()
```

```python
city_group = new_base.groupby(['city', 'type'])
AVR_Fare = city_group.mean()['fare']
Rides_Num = city_group.count()['ride_id']
Drivers_Num = city_group.mean()['driver_count']
#city_type = city_group.sum()

plot_cities = pd.DataFrame({'AVR_Fare': AVR_Fare, 'Rides_Num': Rides_Num,
                            'Drivers_Num': Drivers_Num})

plot_cities.sort_values('Rides_Num', ascending=False).head()
```

```python
#Plot data
import matplotlib.pyplot as plt

plt.scatter(plot_cities['AVR_Fare'], plot_cities['Rides_Num'], s=plot_cities['Drivers_Num'] c=)
plt.show()
```
